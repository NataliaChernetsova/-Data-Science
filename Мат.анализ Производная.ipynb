```python
import numpy as np
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

Уровень 0

Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.

```python
def f(x):
    return x**5 + 4*np.sin(2*x) + np.cos(3*x + 3)
print(f'Значение первой производной функции: {derivative(f, dx = 1e-6, x0=1.0, n=1)}\nЗначение второй производной функции: {derivative(f, dx = 1e-6, x0=1.0, n=2)}')
```

Уровень 1:

Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].


```python
def f(x): 
    return (np.sin(2*x+1))**5
```

```python
x = np.linspace(-5, 5)
fx = f(x)
f1x = [derivative(f, xx, dx = 1e-6, n=1) for xx in x]
f2x = [derivative(f, xx, dx = 1e-6, n=2) for xx in x]
```

```python
plt.figure(figsize=(8,8))
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative1')
plt.plot(x, f2x, label='derivative1')
plt.legend()
plt.show()
```
