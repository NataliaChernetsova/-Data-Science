```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

```python
rv = sts.expon(3)
sample = rv.rvs(size=1000)
```

```python
x = np.linspace(0,4,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
n1 = range(80, 1000, 100)
samples_count = 1000
a=np.zeros((len(n1), samples_count))

for i in range(len(n1)): 
    for number in range(samples_count): 
        sample = rv.rvs(size=n1[i])
        a[i][number] = sample.mean() 
```

```python
fig, (ax0, ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8) = plt.subplots(
    nrows=9, ncols=1,
    figsize=(10, 20)
)

labels=[]
for i in n1:
  labels.append('sample of ' + str(i) + ' histogram')

ax0.hist(a[0], density=True, label=labels[0])
ax1.hist(a[1], density=True, label=labels[1])
ax2.hist(a[2], density=True, label=labels[2])
ax3.hist(a[3], density=True, label=labels[3])
ax4.hist(a[4], density=True, label=labels[4])
ax5.hist(a[5], density=True, label=labels[5])
ax6.hist(a[6], density=True, label=labels[6])
ax7.hist(a[7], density=True, label=labels[7])
ax8.hist(a[8], density=True, label=labels[8])

ax0.set_ylabel('$f(x)$')
ax0.set_xlabel('$x$')
ax0.legend(loc='upper right')

ax1.set_ylabel('$f(x)$')
ax1.set_xlabel('$x$')
ax1.legend(loc='upper right')

ax2.set_ylabel('$f(x)$')
ax2.set_xlabel('$x$')
ax2.legend(loc='upper right')

ax3.set_ylabel('$f(x)$')
ax3.set_xlabel('$x$')
ax3.legend(loc='upper right')

ax4.set_ylabel('$f(x)$')
ax4.set_xlabel('$x$')
ax4.legend(loc='upper right')

ax5.set_ylabel('$f(x)$')
ax5.set_xlabel('$x$')
ax5.legend(loc='upper right')

ax6.set_ylabel('$f(x)$')
ax6.set_xlabel('$x$')
ax6.legend(loc='upper right')


ax7.set_ylabel('$f(x)$')
ax7.set_xlabel('$x$')
ax7.legend(loc='upper right')

ax8.set_ylabel('$f(x)$')
ax8.set_xlabel('$x$')
ax8.legend(loc='upper right')
```
