```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import scipy.stats as sps
```

Задание 1:
Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
def norm(x, mu, sigma):
    norm_rv = sts.norm(loc=mu, scale=sigma)
    return norm_rv.pdf(x)

def expon(x, lam):
    expon_rv = sts.expon(lam)
    return expon_rv.pdf(x)  

def st(x, n):
    t_rv = sts.t(n)
    return t_rv.pdf(x)
```

```python
x = np.linspace(-5,5,100)
plt.plot(x, norm(x,2.0,0.5))
plt.plot(x, norm(x,1.5,0.4))
plt.plot(x, norm(x,1.2,0.3))
plt.plot(x, norm(x,1.8,0.1))
plt.plot(x, norm(x,0.5,0.5))
plt.title('Нормальное распределение')
```

```python
plt.plot(x, expon(x,6.0))
plt.plot(x, expon(x,0.4))
plt.plot(x, expon(x,2.0))
plt.plot(x, expon(x,1.0))
plt.plot(x, expon(x,9.0))
plt.title('Экспоненциальное распределение')
```

```python
plt.plot(x, st(x,10))
plt.plot(x, st(x,3))
plt.plot(x, st(x,6))
plt.plot(x, st(x,1))
plt.plot(x, st(x,0.2))
plt.title('Распределение Стъюдента')
```
